```python
import time
import numpy as np
import matplotlib.pyplot as plt


def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    return C


x = np.arange(1, 9)
n = 2**(x-1) * 3
times_naive = np.zeros(n.size)
times_optimized = np.zeros(n.size)
for i, ni in enumerate(n):
    a = np.random.rand(ni, ni)
    b = np.random.rand(ni, ni)

    time_start = time.perf_counter()
    naive_matrix_multiplication(a, b)
    times_naive[i] = time.perf_counter() - time_start

    time_start = time.perf_counter()
    np.matmul(a, b)
    times_optimized[i] = time.perf_counter() - time_start
    

print("Naive runtimes (s):", times_naive)
print("Optimized runtimes (s):", times_optimized)

plt.subplot(2, 1, 1)
plt.plot(x, times_naive, label='Naive Method')
plt.plot(x, times_optimized, label='Optimized Method')
plt.title('Matrix Number vs. Runtime')
plt.xlabel('Matrix Number')
plt.ylabel('Runtime (s)')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(x, times_naive, label='Naive Method')
plt.plot(x, times_optimized, label='Optimized Method')
plt.yscale('log')
plt.xlabel('Matrix Number')
plt.ylabel('Runtime (s)')
plt.legend()

plt.show()
```
