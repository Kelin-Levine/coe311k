```python
"""Graphing with matplotlib."""

import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0.0, 1.0, 50)
theta = 2.0 * np.pi * t

plt.subplot(2, 2, 1)
plt.plot(theta, np.sin(theta), 'r-o', label='Sin(theta)')
plt.plot(theta, np.cos(theta), 'g-s', label='Cos(theta)')
plt.legend() # Display the legend.
plt.xlabel('Rads') # Add a label to the x-axis.
plt.ylabel('Amplitude') # Add a label to the y-axis.
plt.title('Sin and Cos Waves') # Add a graph title.

plt.subplot(2, 2, 2)
x = np.sin(theta)
y = np.cos(theta)
plt.scatter(x, y, 50.0 * (1.0-t), t)
plt.colorbar()

plt.subplot(2, 2, 3)
x = np.random.rand(2000)
plt.hist(x**2, 50)
plt.plot(t[1:], 6.0 / t[1:], 'r--')

plt.subplot(2, 2, 4)
x = t * 5.0
plt.plot(x, x**3 - 2.0 * x**2, 'b-', label='x^3 - 2x^2')
plt.plot(x, x**2 + 3.0 * x, 'y-', label='x^2 + 3x')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```
