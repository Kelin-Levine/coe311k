```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline


# Revisit the last challenge, add a function that calculates
# the exact solution, and one that calculates the local error
# and the global error for h = .1, .5, 1, 1.5, plot the error


# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y


# Specific solution of above function, initial condition y(0)=1
def f_y(x):
    return np.exp(-2 * x)


# Compute local and global error of solution of y values
def f_error(y_approx, y_exact):
    err = y_approx - y_exact
    return err, np.sum(err)


for h in range(0, 101, 5):
    h = max(h, 1)
    # Define parameters
    h = h / 100.0  # step size
    x_start, x_end = 0, 2
    num_steps = int((x_end - x_start) / h)

    x_values = np.linspace(x_start, x_end, num_steps + 1, dtype=float)
    y_values = np.zeros(num_steps + 1, dtype=float)

    # Initial condition
    y_values[0] = 1.0  # y(0) = 1

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])

    # Calculate exact solution and error
    y_exact = f_y(x_values)
    err_local, err_global = f_error(y_values, y_exact)

    # Plot the results
    plt.subplot(1, 2, 1)
    plt.plot(x_values, y_values, label="Euler's Method")
    plt.plot(x_values, y_exact, label="Exact Solution")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f"Approximation of dy/dx = -2y (step size {h})")
    plt.legend()
    plt.subplot(1, 2, 2)
    plt.plot(x_values, err_local, label="Local Error")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Local Error Plot")
    plt.legend()
    plt.show()

    # Output error
    print("Local error for each point:", err_local)
    print("Global error:", err_global)
```
