```python
import numpy as np
import matplotlib.pyplot as plt


def solve_and_plot(X, Y, i):
    A = np.vstack([X, np.ones(len(X))]).T
    s = np.linalg.lstsq(A, Y, rcond=None)
    m, c = s[0]
    r_sum = s[1][0]
    Y2 = m*X+c

    r = np.abs(Y - Y2)
    ri_max = np.argmax(r)
    ri_min = np.argmin(r)

    plt.figure(figsize=(4,9))
    plt.subplot(3, 1, i)
    plt.scatter(X, Y, c='b')
    plt.plot(X, Y2, 'r-')

    print(f"Part {i}")
    print(f"Slope: {m}, Intercept: {c}")
    print(f"Sum of squared residuals: {r_sum}")
    print(f"Largest residual error: {r[ri_max]} @ ({X[ri_max]},{Y[ri_max]})")
    print(f"Smallest residual error: {r[ri_min]} @ ({X[ri_min]},{Y[ri_min]})\n")


# Part 1
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
solve_and_plot(X, Y, 1)

# Part 2
X = np.sort(np.random.rand(5))
Y = np.random.rand(5)
solve_and_plot(X, Y, 2)

# Part 3
X = np.arange(0.0, 1.0, 0.01)
Y = 2.5*X+0.8 + (2.0*np.random.rand(len(X))-1.0)
solve_and_plot(X, Y, 3)

plt.show()
```
