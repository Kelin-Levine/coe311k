```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_approximate_pi(num_samples: int) -> float:
    """Use the Monte Carlo method with a given number of samples to approximate the value of Pi.

    This implementation is based on the algorithm described in the lecture slides.
    """
    x = np.random.rand(num_samples)
    y = np.random.rand(num_samples)
    return np.sum((x*x + y*y) < 1.0) / num_samples * 4.0

abs_errors = np.empty(100)
rel_errors = np.empty(100)
for i in range(100):
    abs_errors[i] = np.abs(monte_carlo_approximate_pi(1000) - np.pi)
    rel_errors[i] = abs_errors[i] / np.pi

plt.hist(abs_errors, 10, color='r', label='Absolute Error')
plt.hist(rel_errors, 10, color='b', label='Relative Error')
plt.legend()
plt.xlabel('Error')
plt.ylabel('Frequency')
plt.title('Error of Monte Carlo Approximation of Pi')
plt.show()
```
