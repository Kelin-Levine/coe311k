```python
"""Estimating PI with the Monte Carlo Method."""

import random as r

STEPS = 10000000

def approximatePi(numSamples: int) -> float:
    count = 0
    for i in range(numSamples):
        x = r.random()  # The function is symmetric, so range (-1, 1) would be redundant
        y = r.random()
        if x*x + y*y < 1.0:
            count += 1
    return 4.0 * count / numSamples

approximatePi(STEPS)
```

```python
"""Meshes with numpy."""

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f1(x,y):
    return (x*x+y*y)

x = np.linspace(-10, 10, 2000)
y = np.linspace(-10, 10, 2000)
X, Y = np.meshgrid(x, y)
Z = f1(X, Y)

im = plt.imshow(Z, cmap='hot')
fig = plt.colorbar(im,orientation='horizontal')
plt.show(fig)
```
