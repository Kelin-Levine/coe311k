```python
import numpy as np
import matplotlib.pyplot as plt


def plot_df_dx(x, df_dx, name, i):
    plt.subplot(2, 2, i)
    plt.plot(x, np.cos(x), 'r--', label='Exact Derivative')
    plt.plot(x, df_dx, 'b-', label=name)
    plt.xlabel('x')
    plt.ylabel("f'(x)")
    plt.title(f"f'(x): {name}")
    plt.legend()


# Define the function and the domain
f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi  # Interval
N = 50  # Number of sample points

# Create a uniform grid
x = np.linspace(a, b, N)
y = f(x)

# Step size
h = (b - a) / (N - 1)

# Central Difference Approximation for the derivative
df_dx_c = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
df_dx_f = (f(x + h) - f(x)) / h

# Backward Difference Approximation
df_dx_b = (f(x) - f(x - h)) / h

# Plot the true derivative
plt.subplot(2, 2, 1)
plt.plot(x, np.cos(x), 'r-', label='Exact Derivative')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.title("f'(x): Exact Derivative")
plt.legend()

# Plot the approximations
plot_df_dx(x, df_dx_c, 'Central Difference', 2)
plot_df_dx(x, df_dx_f, 'Forward Difference', 3)
plot_df_dx(x, df_dx_b, 'Backward Difference', 4)
plt.show()

## Do a canvas of 4 subplots or 4 graphs
## on one plot comparing cos(x), Forward
## Difference, Backwards Difference, and
## Central Difference Methods
```
