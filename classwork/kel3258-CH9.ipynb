```python
import numpy as np
import matplotlib.pyplot as plt


def solve_and_plot(X, Y, i):
    A = np.vstack([X, np.ones(len(X))]).T
    m, c = np.linalg.lstsq(A, Y, rcond=None)[0]
    Y2 = m*X+c

    r = Y - Y2
    r_abs = np.abs(r)
    ri_max = np.argmax(r_abs)
    ri_min = np.argmin(r_abs)

    plt.figure(figsize=(4,9))
    plt.subplot(3, 1, i)
    plt.scatter(X, Y, c='b')
    plt.plot(X, Y2, 'r-')

    print(f"Slope: {m}, Intercept: {c}")
    print(f"Residuals: {r}")
    print(f"Largest residual error: {r_abs[ri_max]} @ ({X[ri_max]}, {Y[ri_max]})")
    print(f"Smallest residual error: {r_abs[ri_min]} @ ({X[ri_min]}, {Y[ri_min]})\n")


# Part 1: Given Data
print(f"Part 1: Given Data")
X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])
solve_and_plot(X, Y, 1)
plt.title("Part 1: Given Data")

# Part 2: Random Vectors
print("Part 2: Random Vectors")
X = np.sort(np.random.rand(5))
Y = np.random.rand(5)
solve_and_plot(X, Y, 2)
plt.title("Part 2: Random Vectors")

# Part 3: Linear Function with Error
print("Part 3: Linear Function with Error")
X = np.arange(0.0, 1.0, 0.01)
Y = 2.5*X+0.8 + (2.0*np.random.rand(len(X))-1.0)
solve_and_plot(X, Y, 3)
plt.title("Part 3: Linear Function with Error")

plt.show()
```
