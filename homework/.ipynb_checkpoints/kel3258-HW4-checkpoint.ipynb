```python
import numpy as np
import matplotlib.pyplot as plt


def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    errs = []
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        # Check for convergence
        err = np.linalg.norm(x_new - x, ord=np.inf)
        errs.append(err)
        if err < tol:
            return x_new, k+1, errs
        x = x_new
    return x, max_iterations, errs


def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    errs = []
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        # Check for convergence
        err = np.linalg.norm(x_new - x, ord=np.inf)
        errs.append(err)
        if err < tol:
            return x_new, k+1, errs
        x = x_new
    return x, max_iterations, errs


def compute_and_plot(A, b, tol, max_iterations, i):
    x0 = np.zeros_like(b)
    _, j_iters, j_errs = jacobi(A, b, x0, tol, max_iterations)

    x0 = np.zeros_like(b)
    _, gs_iters, gs_errs = gauss_seidel(A, b, x0, tol, max_iterations)

    print(f"Tolerance: {tol}")
    print(f"Jacobi Iterations: {j_iters}")
    print(f"Gauss-Seidel Iterations: {gs_iters}\n")

    plt.figure(figsize=(4,9))
    plt.subplot(3, 1, i)
    # plt.subplot(3, 2, i*2-1)
    plt.plot(range(1, j_iters+1), j_errs, 'r-', label='Jacobi')
    plt.plot(range(1, gs_iters+1), gs_errs, 'b-', label='Gauss-Seidel')
    plt.legend()
    plt.title(f"Convergence; Tolerance: {tol}")
    plt.xlabel('Iteration')
    plt.ylabel('Absolute Error to Previous')

    # plt.figure(figsize=(4,9))
    # plt.subplot(3, 2, i*2)
    # plt.plot(range(j_iters-1, j_iters+1), j_errs[-2:], 'r-', label='Jacobi')
    # plt.plot(range(gs_iters-1, gs_iters+1), gs_errs[-2:], 'b-', label='Gauss-Seidel')
    # plt.legend()
    # plt.xlabel('Iteration')


# Parameters
max_iterations = 100
A = np.array([[3, -0.1, -0.2],
              [0.1, 7, -0.3],
              [0.3, -0.2, 10]])
b = np.array([7.85, -19.3, 71.4])

# Compute and Plot Solutions
compute_and_plot(A, b, 1e-3, max_iterations, 1)
compute_and_plot(A, b, 1e-6, max_iterations, 2)
compute_and_plot(A, b, 1e-9, max_iterations, 3)
plt.show()

# Decreasing tolerance increases the number of iterations needed for convergence
# because a lower tolerance requires a smaller error to consider the method to have
# converged. We can see in the charts of convergence for this system of equations
# that absolute error is always decreasing with each iteration, so to reach a smaller
# error for a smaller tolerance requires more iterations. Additionally, the Jacobi
# Method tends to require more iterations to converge than the Gauss-Seidel Method,
# and this becomes more apparent at smaller tolerances.
```
