# Review of Part 2

In this project, continuing on the problem exploring in Part 2, I will solve an
overdamped spring-mass system. This system describes how the displacement of a
mass on a highly damped spring oscillates as it experiences an external force.
This is a stiff equation, which means that it will change very rapidly for a
short period of time in the beginning, and then it will change slowly over a
much longer period of time.

$$\frac{dx}{dt}=-5000x+0.002\sin(t)$$

Previously, in Part 2, this system was solved using a basic implicit solver
based on Euler’s Backward Method with a Newton-Raphson solver. This works by
using an iterative solver to step through the solution, iteratively guessing
the solution at the next step that, based on the rate of change, connects back
to the previous step.

# Newton-Raphson Damping Implementation

The basic implicit solver used previously can fail when the initial guess is far
from the solution, the step size is very large, the function is highly nonlinear
in certain regions, or when the update overshoots and diverges. It uses this
formula for the Newton-Raphson update formula:

$$x_{i+1} = x_i - \frac{g(x_i)}{g'(x_i)}$$

Where $x_i$ is the previous guess for displacement at the next step, $x_{i+1}$
is the next guess after the iteration, and $g$ is the residual function.

In this part of the project, I will use Damped Newton-Raphson to solve the
problem. This adds a damping factor $\alpha\in (0,1]$ to the Newton-Raphson
update formula:

$$x_{i+1} = x_i - \alpha\frac{g(x_i)}{g'(x_i)}$$

where $\alpha=1$ takes the full step size and $\alpha<1$ takes smaller, safer
steps when convergence is difficult.

The damping strategy is to do a line search that chooses a value for $\alpha$
that ensures the residual decreases with each iteration. We only accept steps
that reduce the residual.

```python
"""Setup Numerical Implementation

This cell must be run before any following numerical implementation. This is to
load the necessary parameters and functions into the kernel. It should not
produce any output by itself.

For convenience, this also includes a function to plot solutions of the ODE.
"""


# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt


# Define the parameters
t_i = 0.0  # Start of interval
t_f = 2.0  # End of interval
x_i = 1.0  # Initial condition of x

h_values_implicit = [2.0, 0.1, 0.01, 0.001, 0.0001]  # Step sizes (implicit method)
h_adaptive = 0.0001  # Initial step size (adaptive method)


# Define the ODE function
def f(t, x):
    """Function representing the dx/dt ODE"""
    return -5000*x + 0.002*np.sin(t)


# Convenience function for plotting a solution to the ODE
def plot_solution(solver, title, h_values):
    # Configure plot
    plt.title(f'Displacement of Mass Using {title}')
    plt.xlabel('Time (s)')
    plt.ylabel('Displacement (m)')
    plt.grid(True)

    # Run the solver, plot and record all solutions
    all_values = []
    for h in h_values:
        values = solver(f, x_i, t_i, t_f, h)
        plt.plot(values[0], values[1], label=f'Step size: h={h}')
        all_values.append(values)

    # Finish plotting
    plt.legend(loc='upper left')

    # Return solutions (for optional processing)
    return all_values
```

```python
"""Damped Newton-Raphson Root-Finding Algorithm

Implementations of Damped Newton-Raphson Method to predict displacement at the
next time step using the residual function.

This includes three implementations of Newton-Raphson: a general version that
takes any residual function, an "auto" version that implements the residual
function derived in 'Mathematical Setup' into the general implementation, and an
"auto optimized" version for this residual function with a unique implementation
to take advantage of its simplified update formula.

The residual function is programmed into relevant implementations directly and
is NOT based on the ODE function in the 'Setup Numerical Implementation' cell.

This cell runs after the 'Setup Numerical Implementation' cell in the kernel.
"""


# Damped Newton-Raphson Implementation
def damped_newton_raphson(g, dg, x_i, tol=1e-6, max_iter=20, alpha_min=1e-4):
    # Set initial guess
    x = x_i
    flag_damping = False
    # Newton-Raphson iteration loop
    for i in range(max_iter):
        # Update guess using Damped Newton-Raphson formula
        g_val = g(x)
        delta = -g_val / dg(x)
        x_new = x + delta
        # Finish if residual is within tolerance
        if abs(g_val) < tol:
            return x_new, i + 1, flag_damping
        
        # Line search for damping factor
        alpha = 1.0
        g_current = abs(g_val)
        for _ in range(10):  # Backtrack: max 10 steps
            x_newer = x + alpha * delta
            g_new = abs(g(x_newer))
            if g_new < g_current:  # Sufficient decrease
                x_new = x_newer
                flag_damping = True
                break
            alpha = alpha / 2.0
            if alpha < alpha_min:
                print(f"Warning: alpha too small at iter {i+1}")
                break

        x = x_new
    # Failure: residual was never within tolerance
    return None, max_iter, False


def gen_g(t_n1, x_n, h):
    """Generates an anonymous function representing g(x_n+1)."""
    phi1 = 1.0 + 5000.0*h
    phi2 = x_n + 0.002*h*np.sin(t_n1)
    return lambda x_n1: phi1*x_n1 - phi2


def gen_dg(h):
    """Generates an anonymous function representing g'(x_n+1)."""
    phi = 1.0 + 5000.0*h
    return lambda x_n1: phi


# Newton-Raphson (for the specific system's residual function)
def damped_newton_raphson_auto(t_n1, x_n, h, **kwargs):
    # Obtain residual function
    g = gen_g(t_n1, x_n, h)
    dg = gen_dg(h)
    # Newton-Raphson (using current value as initial guess)
    return damped_newton_raphson(g, dg, x_n, **kwargs)


# Newton-Raphson (for the specific system's residual function, optimized)
def damped_newton_raphson_auto_optimized(t_n1, x_n, h, **kwargs):
    # Iterating is redundant because previous guess is not used by formula
    x_new = (x_n + 0.002*h*np.sin(t_n1)) / (1 + 5000*h)
    return x_new, 1, False
```

```python
"""Euler's Backward Method with Newton-Raphson

Solve the first-order differential equation using Euler's Backward Method with
Newton-Raphson to solve the residual function.

This cell runs after the 'Damped Newton-Raphson Root-Finding Algorithm' cell
in the kernel.
"""


# Euler's Backward Method Implementation
def backward_euler(f, x_i, t_i, t_f, h, nr=damped_newton_raphson_auto):
    # Create arrays to store t and x values
    t_values = np.arange(t_i, t_f + h, h)
    x_values = np.zeros_like(t_values)
    iterations = np.zeros(t_values.size-1)
    damping_needed = np.zeros(t_values.size-1, dtype=bool)

    # Set initial condition
    x_values[0] = x_i

    # Euler's Backward Method step loop
    for i in range(1, len(t_values)):
        # Current values
        t_n1 = t_values[i]
        x_n = x_values[i - 1]
        # Update solution
        x_n1, max_iter, was_damping_needed = nr(t_n1, x_n, h)
        # Record values
        if x_n1 is None:
            print(f"Warning: failed to find roots for t={t_n1} !")
        x_values[i] = x_n1
        iterations[i-1] = max_iter
        damping_needed[i-1] = was_damping_needed
    return t_values, x_values, iterations, damping_needed


# Optional: use the optimized NR implementation
# - Both give the same value for x_n1; default may have greater truncation error
# - Default may complete in 1 or 2 iters; optimized always completes in 1
def backward_euler_optimized(f, x_i, t_i, t_f, h):
    return backward_euler(f, x_i, t_i, t_f, h, nr=damped_newton_raphson_auto_optimized)
```

```python
"""Plot Euler's Backward Method with Newton-Raphson

Plot the solution of the differential equation using Euler's Backward Method
with Newton-Raphson to solve the residual function.

This cell runs after the 'Euler's Backward Method with Newton-Raphson' cell in
the kernel.
"""


# Setup plot
plt.figure(figsize=(10, 10))

# Solve and plot solutions
plt.subplot(2, 1, 1)
values_implicit = plot_solution(backward_euler, "Euler's Backward Method (Damped NR)", h_values_implicit)
plt.subplot(2, 1, 2)
plot_solution(backward_euler_optimized, "Euler's Backward Method (Damped NR, Optimized)", h_values_implicit)

# Show plot
plt.show()

# Report stats
x_final = [v[1][-1] for v in values_implicit]
num_timesteps = [len(v[2]) for v in values_implicit]
avg_iters = [np.mean(v[2]) for v in values_implicit]
num_funcs = [sum(v[2]) for v in values_implicit]
needed_damping = [np.mean(v[3]) for v in values_implicit]
print("Stats of the solutions using the full non-optimized Newton-Raphson solver:")
for i, h in enumerate(h_values_implicit):
    print(f"\nSolution of step size h={h}:")
    print(f"Final value: x={x_final[i]}")
    print("Total number of time steps:", num_timesteps[i])
    print("Average Newton-Raphson iterations per step:", avg_iters[i])
    print("Total number of Newton-Raphson function evaluations:", num_funcs[i])
    print(f"Percentage of steps that needed damping: {needed_damping[i]*100:.5}%")
```

This implementation uses Damped Newton-Raphson to solve the differential equation.

Unfortunately, for this specific system, it is not possible to demonstrate test cases
where basic Newton-Raphson fails but Damped Newton-Raphson succeeds. For reasons
discussed in Part 2, the previous guess cancels out in the Newton-Raphson update
formula for this system. Basic Newton-Raphson will always find the same solution in
every iteration, and thus it is mathematically impossible for basic Newton-Raphson
to fail for this specific system. In reality, basic Newton-Raphson could still fail for
computational reasons, such as truncation error, but this would only occur with very
unreasonable parameters.

Nevertheless, Damped Newton-Raphson is more reliable and less likely to fail than
basic Newton-Raphson, especially with very large initial step sizes or highly
nonlinear residual functions.

# Adaptive Step Size Implementation

This implementation can be improved further: the solution may change rapidly in
some regions, but slowly in others. A fixed step size can be too large and
produce large errors in difficult regions while also being too small and wasting
computation in easy regions. Adaptive step size automatically adjusts step size
on the fly to maintain target accuracy with minimal computation.

The step-doubling approach achieves this by comparing a big step to two smaller
steps. We take one step of the full size, solve it with Newton-Raphson, then
take two steps at half size, solve them with Newton-Raphson, and then estimate
the local error using the difference between the two approaches:

$$\text{err}=\left| x_{n+1}^{(h/2)} - x_{n+1}^{(h)} \right|$$

```python
"""Euler's Backward Method with Adaptive Step Size

Solve the first-order differential equation using Euler's Backward Method with
Newton-Raphson to solve the residual function and adaptive step sizes.

This cell runs after the 'Damped Newton-Raphson Root-Finding Algorithm' cell
in the kernel.
"""


# Adaptive Euler's Backward Method Implementation
def backward_euler_adaptive(f, x_i, t_i, t_f, h_init, nr=damped_newton_raphson_auto, tol=1e-6):
    t_values = [t_i]
    x_values = [x_i]
    h_values = [h_init]
    err_values = []
    t = t_i
    x = x_i
    h = h_init
    i = 0
    nr_iters = 0
    while t < t_f:
        i += 1
        if t + h > t_f:  # Adjust step size to not overshoot
            h = t_f - t
        # Single step with h
        x_full, iters, _ = nr(t+h, x, h, tol=tol)
        nr_iters += iters
        # Two half-steps with h/2
        h_half = h / 2
        x_half_1, iters, _ = nr(t + h_half, x, h_half, tol=tol)
        nr_iters += iters
        x_half_2, iters, _ = nr(t + h, x_half_1, h_half, tol=tol)
        nr_iters += iters
        # Error estimation
        error = abs(x_full - x_half_2)
        # Adjust step size
        if error > tol:
            h /= 2  # Decrease step size
            continue  # Retry the step with smaller h
        elif error < tol / 2:
            h *= 2  # Increase step size for efficiency
        # Accept the step
        t += h
        x = x_half_2
        t_values.append(t)
        x_values.append(x)
        h_values.append(h)
        err_values.append(error)

    return np.array(t_values), np.array(x_values), np.array(h_values), i, nr_iters, err_values


# Adaptive Euler's Backward Method Implementation (optimized NR)
def backward_euler_adaptive_optimized(f, x_i, t_i, t_f, h_init, tol=1e-6):
    return backward_euler_adaptive(f, x_i, t_i, t_f, h_init, tol=tol, nr=damped_newton_raphson_auto_optimized)
```

```python
"""Plot Euler's Backward Method with Adaptive Step Size

Plot the solution of the differential equation using Euler's Backward Method
with Newton-Raphson to solve the residual function and adaptive step sizes.

This cell runs after the 'Euler's Backward Method with Adaptive Step Size' cell
in the kernel.
"""


def adaptive_triple_plot(solver, title, nrows, i, h=h_adaptive, tol=1e-6):
    # Configure first subplot
    plt.subplot(nrows, 3, i*3-2)
    plt.title(f'Displacement of Mass Using {title}')
    plt.xlabel('Time (s)')
    plt.ylabel('Displacement (m)')
    plt.grid(True)
    # Solve and plot solution
    values = solver(f, x_i, t_i, t_f, h, tol=tol)
    plt.plot(values[0], values[1], 'o-', label=f'Initial step size: h={h}')
    plt.legend(loc='upper left')

    # Configure second subplot
    plt.subplot(nrows, 3, i*3-1)
    plt.title(f'Step Size vs. Time Using {title}')
    plt.xlabel('Time (s)')
    plt.ylabel('Step size')
    plt.grid(True)
    # Plot h vs. time
    plt.plot(values[0], values[2], 'o-')

    # Configure third subplot
    plt.subplot(nrows, 3, i*3)
    plt.title(f'Error Estimate vs. Time Using {title}')
    plt.xlabel('Time (s)')
    plt.ylabel('Error estimate')
    plt.grid(True)
    # Plot error estimate vs. time
    plt.plot(values[0][1:], values[5], 'o-')
    return values


def adaptive_report_stats(values):
    print(f"Final value: x={values[1][-1]}")
    print("Total number of iterations:", values[3])
    print("Total number of Newton-Raphson function evaluations:", values[4])


# Setup plot
plt.figure(figsize=(25, 10))

# Solve and plot solutions
values_adaptive = adaptive_triple_plot(backward_euler_adaptive, "Euler's Backward Method (Adaptive h)", 2, 1)
adaptive_triple_plot(backward_euler_adaptive_optimized, "Euler's Backward Method (Adaptive h, Optimized)", 2, 2)

# Show plot
plt.show()

# Report stats
print(f"Solution of step size h={h_adaptive}:")
adaptive_report_stats(values_adaptive)
```

# Conclusions

In conclusion, depending on the equation, damping can greatly benefit the
robustness of the solution, and adaptive step sizes greatly improve both the
robustness and the efficiency of the solution without compromising on accuracy.
Damped Newton-Raphson is able to handle higher step sizes, farther initial
guesses, and stiffer systems than basic Newton-Raphson by damping the guess to
automatically take smaller, safer steps as necessary. Adaptive step sizes can
greatly improve the efficiency of the method by automatically reducing overall
step size for portions of the solution that do not require it. This is especially
effective for stiff equations, where part of the solution is changing
significantly faster than other parts. However, both damping and adaptive step
sizes increase the complexity of each step, so for systems that do not benefit
from them, they may be less efficient than simpler methods. Overall, damping and
adaptive step sizes are very effective at improving the robustness and efficiency
of solutions to stiff equations without losing accuracy, but should only be used
when these benefits outweigh the added complexity.
