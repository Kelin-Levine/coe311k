```python
"""Homework 1 - Plot Monte Carlo Pi

Iterate the number of samples for a Monte Carlo Pi approximation from 10 to 10000, in
steps of 20, and plot the error (absolute value of the calculated value of Pi minus
the actual value of Pi encoded in Python, to 15 digits past the decimal point).

Kelin Levine
February 3, 2026
COE 311K - Engineering Computation, Spring 2026
The University of Texas at Austin
AI was not used in the creation of this code.
"""


import numpy as np
import matplotlib.pyplot as plt


def monte_carlo_approximate_pi(num_samples: int) -> float:
    """Use the Monte Carlo method with a given number of samples to approximate the value of Pi.

    This implementation is based on the algorithm described in the lecture slides.
    """
    x = np.random.rand(num_samples)
    y = np.random.rand(num_samples)
    return np.sum((x*x + y*y) < 1.0) / num_samples * 4.0


# Compute errors of approximations from 10 to 10000, in steps of 20
approx_range = np.arange(10, 10000, 20)
errors = np.zeros(len(approx_range))
for i, v in enumerate(approx_range):
    errors[i] = np.abs(monte_carlo_approximate_pi(v) - np.pi)

# Plot error vs number of samples
plt.plot(approx_range, errors, 'r-')
plt.xlabel('Number of samples')
plt.ylabel('Error')
plt.title('Error of Monte Carlo Approximation of Pi')
plt.show()
```
